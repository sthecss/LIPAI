A distribuição cumulativa empírica é uma função que explica como valores associadas a uma probabilidade. Dados os exemplos, o primeiro gráfico, temos que SepalLength (Comprimento da sépala) tem 40% dos dados tem 5,5 cm. Essa ferramenta traz um panorama empírico da característica.

```python
# Empirical Cumulative Distribution
# visualising ECDF

from mlxtend.plotting import ecdf
import matplotlib.pyplot as plt

# Criando uma figura e eixos para os subplots
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(13, 40))

# Ajustando o espaçamento entre os subplots
plt.subplots_adjust(right=2)
plt.subplots_adjust(top=2)

# Limpando a figura
plt.clf()

# Iterando sobre cada coluna do dataset
for i, feature in enumerate(dataset.columns, 1):
    # Adicionando um subplot para cada coluna
    plt.subplot(len(dataset.columns), 3, i)

    # Criando um gráfico de funções de distribuição cumulativa
    ecdf(dataset[feature])

    # Adicionando título ao subplot
    plt.title(f'{feature}', size=15, fontsize=12, fontweight='medium')

    # Removendo as grades do gráfico
    plt.grid(False)

    # Ajustando o tamanho das etiquetas nos eixos x e y
    plt.tick_params(axis='x', labelsize=8)
    plt.tick_params(axis='y', labelsize=8)

# Ajustando o layout dos subplots
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

3) O dataset Iris não possui valores ausentes. A base escolhida também pode conter todos os descritores por cada amostra. Use um dataset e sobre as amostras impute valores ausentes usando a média, mediana ou outras medidas.

```python
# Verificando valores ausentes
print("Valores ausentes por coluna:")
print(dataset.isnull().sum())
print(f"\nTotal de amostras com pelo menos 1 ausente: {dataset.isnull().any(axis=1).sum()}")
```

```python
# Exibindo as linhas com valores ausentes
dataset.loc[dataset.isnull().any(axis=1)]
```

```python
# Imputação: média para bill_length e flipper_length, mediana para bill_depth e body_mass
dataset_imputed = dataset.copy()

dataset_imputed['bill_length_mm']    = dataset_imputed['bill_length_mm'].fillna(dataset_imputed['bill_length_mm'].mean())
dataset_imputed['flipper_length_mm'] = dataset_imputed['flipper_length_mm'].fillna(dataset_imputed['flipper_length_mm'].mean())
dataset_imputed['bill_depth_mm']     = dataset_imputed['bill_depth_mm'].fillna(dataset_imputed['bill_depth_mm'].median())
dataset_imputed['body_mass_g']       = dataset_imputed['body_mass_g'].fillna(dataset_imputed['body_mass_g'].median())

print("Valores ausentes após imputação:")
print(dataset_imputed.isnull().sum())
```

4) Após aplicação das operações, faça a representação gráfica com histogramas das novas características dos dados analisados.

```python
fig, axes = plt.subplots(4, 2, figsize=(14, 16))
fig.suptitle('Antes vs Depois da Imputação', fontsize=14)

for i, col in enumerate(cols):
    # antes
    axes[i, 0].hist(dataset[col].dropna(), bins=20, color='steelblue', edgecolor='white')
    axes[i, 0].set_title(f'{col} — Original')
    axes[i, 0].set_ylabel('Frequência')
    # depois
    axes[i, 1].hist(dataset_imputed[col], bins=20, color='seagreen', edgecolor='white')
    axes[i, 1].set_title(f'{col} — Após Imputação')

plt.tight_layout()
plt.show()
```

6) No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python
fractions = {'10%': 0.10, '30%': 0.30, '50%': 0.50}
samples = {}

for label, frac in fractions.items():
    samples[label] = dataset_imputed.sample(frac=frac, random_state=42)
    print(f"Amostra {label}: {len(samples[label])} amostras")
```

```python
# Comparação das métricas entre as amostras e o dataset completo
print(f"{'Métrica':<22}", end='')
print(f"{'Completo':>12}", end='')
for label in fractions:
    print(f"{label:>12}", end='')
print()
print('-' * 60)

for col in cols:
    for metric, fn in [('Média', 'mean'), ('Mediana', 'median'), ('Desvio Padrão', 'std')]:
        print(f"{col[:10]} {metric:<12}", end='')
        print(f"{getattr(dataset_imputed[col], fn)():>12.2f}", end='')
        for label in fractions:
            print(f"{getattr(samples[label][col], fn)():>12.2f}", end='')
        print()
    print()
```

```python
# Visualização: médias por amostra
means_full = dataset_imputed[cols].mean()
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Média por Atributo — Comparação de Amostras', fontsize=13)

for ax, col, title in zip(axes.flat, cols, titles):
    values = [dataset_imputed[col].mean()] + [samples[l][col].mean() for l in fractions]
    labels = ['100%'] + list(fractions.keys())
    colors = ['steelblue', 'gold', 'tomato', 'seagreen']
    ax.bar(labels, values, color=colors, edgecolor='white')
    ax.set_title(title)
    ax.set_ylabel('Média')
    ax.set_ylim(min(values) * 0.97, max(values) * 1.03)

plt.tight_layout()
plt.show()

print("""
Observações sobre a amostragem progressiva:
- Com 10% (~34 amostras), as métricas já se aproximam do dataset completo,
  mas desvios maiores são esperados pela menor representatividade.
- Com 30% (~103 amostras), as médias e medianas ficam bem próximas do total.
- Com 50% (~172 amostras), os resultados são praticamente idênticos ao dataset
  completo — a Lei dos Grandes Números começa a se consolidar.
- Atributos com maior variância (body_mass_g) são os mais sensíveis a amostras
  pequenas.
""")
```
