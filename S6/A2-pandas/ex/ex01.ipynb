## 2 - Identifique em quais imagens o modelo errou a predição.

```python
w_pred = (df["real_class"] != df["predicted_class"]).sum()
print(w_pred)
```

## 3 - Verifique se o modelo estava confiante mesmo quando errou.

```python
w_pred = df[df["real_class"] != df["predicted_class"]]
```

```python
w_pred
```

```python
w_malign = w_pred[w_pred["predicted_class"] == 'malign']
w_bening = w_pred[w_pred["predicted_class"] == 'benign']

prob_malign = w_malign['prob_malign']
prob_bening = w_bening['prob_benign']

probs_df = pd.concat([prob_malign, prob_bening])

mean_trust = probs_df.mean()

print(f"Confiança média do modelo ao errar foi: {mean_trust:.2f}")
```

### 4 -Calcule:
a. TP (real malign, previsto malign)

b. TN (real benign, previsto benign)

c. FP (real benign, previsto malign)

d. FN (real malign, previsto benign)

```python
df
```

```python
malign_df = df[df['real_class'] == 'malign']
benign_df = df[df['real_class'] == 'benign']

TP = (malign_df["real_class"] == malign_df['predicted_class']).sum()

TN = (benign_df["real_class"] == benign_df['predicted_class']).sum()

FP = (benign_df["real_class"] != benign_df['predicted_class']).sum()

FN = (malign_df["real_class"] != malign_df['predicted_class']).sum()

print(f"""
TP -> {TP}
TN -> {TN}
FP -> {FP}
FN -> {FN}
""")
```

### 5 - Calcule:
a. Acurácia: (TP+TN)/(TP+TN+FP+FN)

b. Precisão: TP/(TP+FP)

c. Recall: TP/(TP+FN)

d. Especificidade: TN/(TN+FP)

```python
acc = (TP+TN)/(TP+TN+FP+FN)

prec = TP/(TP+FP)

recall = TP/(TP+FN)

espec = TN/(TN+FP)

print(f"""
Acurácia\t -> {acc:.2f}
Precisão\t -> {prec:.2f}
Recall\t\t -> {recall:.2f}
Especificidade\t -> {espec:.2f}
""")
```

### 6 - Encontre as 5 imagens benign com menor prob_benign. O que isso pode indicar?

```python
benign_df = df[df['real_class'] == 'benign'].sort_values(by='prob_benign')

lower_5_images = benign_df[:5]
lower_5_images
```

Mostra indicios de que a probabilidade do modelo dar um falso negativo é grande, já que a probabilidade prevista de ser benigno é pequena

### 7 - Encontre as 5 imagens malign com maior prob_benign. O que isso pode indicar?

```python
malign_df = df[df['real_class'] == 'malign'].sort_values(by='prob_benign', ascending=False)
malign_df
higher_5_images = malign_df[:5]
higher_5_images
```

Mostra indicios de que a probabilidade de dar um falso positivo é grande, já que a probabilidade prevista de ser benigno é grande
